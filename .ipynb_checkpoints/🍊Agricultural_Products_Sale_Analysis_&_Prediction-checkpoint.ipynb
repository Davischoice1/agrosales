Agricultural economics is important because it helps to understand the agricultural economy, which is a key part of the global economy. Agricultural economics studies how resources are allocated, distributed, and used in farming, and how commodities are produced. It also looks at how environmental and technological factors impact yield and demand.

Agriculture is the practise of raising animals and growing crops for food like fruits, vegetables, ,field crops and other products. While agriculture creates jobs and helps in economic development, it also contributes to international trade, countries export agricultural products, it provides food security by supplying agricultural products.

This dataset provides comprehensive information on agricultural product sales, supply chain metrics, and pricing across various categories, including Vegetables, Fruits, Grains, Dairy, and Livestock, for the years 2022 and 2023, which can be used for analysis of agricultural products and for prediction of quantities sold.

# <span style="color:brown">Train Test Split</span>

```python
X = df.drop(['units_sold_kg', 'category', 'supplier', 'farm_location'],  axis = 1)
y = df['units_sold_kg']
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 40)
```

# <span style="color:olivedrab">Data Preprocessing</span>

```python
categorical_col = [cname for cname in X_train.columns if X_train[cname].dtype == 'object']
numerical_col = [cname for cname in X_train.columns if X_train[cname].dtype in ['int', 'float']]
```

```python
print('Categorical Columns: ', categorical_col)
print('Numerical Columns: ', numerical_col)
```

```python
categorical_transformer = Pipeline(steps = [
    ('ordinal', OrdinalEncoder(handle_unknown = 'use_encoded_value', unknown_value = -1))
])
numerical_transformer = Pipeline(steps = [
    ('scaler', StandardScaler())
])

preprocessor = ColumnTransformer([
    ('cat', categorical_transformer, categorical_col),
    ('num', numerical_transformer, numerical_col)
])
```

# <span style="color:peachpuff">Model Building</span>

# <span style="color:aqua">1. Linear Regression</span>

```python
import joblib

linear = LinearRegression()
linear_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', linear)
])
linear_pipeline.fit(X_train, y_train)
joblib.dump(linear_pipeline, 'linear_model.pkl')
```

```python
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

linear_pred = linear_pipeline.predict(X_test)

linear_r2 = r2_score(y_test, linear_pred)
linear_rmse = np.sqrt(mean_squared_error(y_test, linear_pred))

# Residual std for linear
linear_train_pred = linear_pipeline.predict(X_train)
linear_residual_std = np.std(y_train - linear_train_pred)
print(f"Linear Model Residual Std: {linear_residual_std:.2f}")

print("R¬≤ Score:", linear_r2)
print("Root Mean Squared Error (RMSE):", linear_rmse)
```

# <span style="color:indigo">2. Random Forest Regressor</span>

```python
forest = RandomForestRegressor(n_estimators = 200, random_state = 2)
forest_pipeline = Pipeline(steps = [
    ('preprocessor', preprocessor),
    ('model', forest)
])

forest_pipeline.fit(X_train, y_train)
joblib.dump(forest_pipeline, 'forest_model.pkl')
```

```python
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

# Predict with your Random Forest pipeline
forest_pred = forest_pipeline.predict(X_test)

# Evaluate the model
forest_r2 = r2_score(y_test, forest_pred)
forest_rmse = np.sqrt(mean_squared_error(y_test, forest_pred))

# Print results
print("R¬≤ Score:", forest_r2)
print("Root Mean Squared Error (RMSE):", forest_rmse)
```

# <span style="color:lightgreen">3. Gradient Boosting Regressor</span>

```python
gradient = GradientBoostingRegressor(n_estimators=500, learning_rate=0.2, random_state=2)
gradient_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', gradient)
])
gradient_pipeline.fit(X_train, y_train)
joblib.dump(gradient_pipeline, 'gradient_model.pkl')
```

```python
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

# Predict using the gradient boosting pipeline
gradient_pred = gradient_pipeline.predict(X_test)

# Calculate metrics
gradient_r2 = r2_score(y_test, gradient_pred)
gradient_rmse = np.sqrt(mean_squared_error(y_test, gradient_pred))

# Display results
print("R¬≤ Score:", gradient_r2)
print("Root Mean Squared Error (RMSE):", gradient_rmse)
```

# <span style="color:sienna">Model Comparison</span>

```python
comparison = pd.DataFrame()

comparison['Type'] = ['Linear Regression', 'Random Forest Regressor',
                      'Gradient Boosting Regressor']
comparison['r2 Score'] = [linear_r2, forest_r2, gradient_r2]
comparison['Root Mean Squared Error'] = [linear_rmse, forest_rmse, gradient_rmse]

comparison
```

```python
fig = make_subplots(rows=1, cols=2, subplot_titles=['R¬≤ Score', 'RMSE'], horizontal_spacing=0.2)

fig.add_trace(go.Bar(x=comparison['Type'], y=comparison['r2 Score'], name='R¬≤ Score',
                     text=comparison['r2 Score']), row=1, col=1)

fig.add_trace(go.Bar(x=comparison['Type'], y=comparison['Root Mean Squared Error'], name='RMSE',
                     text=comparison['Root Mean Squared Error']), row=1, col=2)

fig.update_traces(texttemplate='%{text:.3f}', marker_line_color='salmon',
                  marker_color=['mintcream', 'cornsilk', 'linen'], marker_line_width=1)

fig.update_layout(title={'text': '<b>Model Comparison</b>', 'x': 0.5, 'xanchor': 'center'},
                  font_color='orchid', showlegend=False,
                  xaxis=dict(title='Model'), xaxis2=dict(title='Model'),
                  yaxis=dict(title='R¬≤ Score'),
                  yaxis2=dict(title='RMSE', type='log'))

fig.show()
```

After comparing the models we find Linear Regression performs best with r2 score of 100% and RMSE of 0.00.
